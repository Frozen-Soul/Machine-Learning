```python
df = df[df['value']>=0]
```

```python
df.head()
```

```python
len(df)
```

```python
sns.catplot(x=df["location"], y=df["value"], hue=df["parameter"], kind="swarm", data=df)
```

```python
plt.style.use('seaborn-darkgrid')
 
# create a color palette
palette = plt.get_cmap('Set1')
 
# multiple line plot
num=0
for column in df.drop('parameter', axis=1):
    num+=1
    plt.plot(df['parameter'], df[column], marker='', color=palette(num), linewidth=1, alpha=0.9, label=column)

    
# Add legend
plt.legend(loc=2, ncol=2)
 
# Add titles
plt.title("A (bad) Spaghetti plot", loc='left', fontsize=12, fontweight=0, color='orange')
plt.xlabel("Time")
plt.ylabel("Score")
```
